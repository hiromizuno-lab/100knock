## ノック15:金額欠損値を補完しよう

```python
uriage.isnull().sum()
```

```python
flag_is_null = uriage['item_price'].isnull()
flag_is_null
```

```python
uriage.loc[flag_is_null, 'item_name'].unique()
```

```python
for trg in list(uriage.loc[flag_is_null, 'item_name'].unique()):
    # item_nameが同じものかつNullではないもののうち、最大の価格を所得
    price = uriage.loc[(~flag_is_null) & (uriage['item_name'] == trg), 'item_price'].max()
    # 取得したpriceを代入する
    uriage['item_price'].loc[(flag_is_null) & (uriage['item_name'] == trg)] = price
```

```python
uriage.isnull().sum()
```

```python
uriage.loc[uriage['item_name'] == '商品A']["item_price"].max()
```

```python
for trg in list(uriage['item_name'].sort_values().unique()):
    print(trg + 'の最大額' + str(uriage.loc[uriage['item_name'] == trg]['item_price'].max()) + '  ' +
          trg + 'の最小額' + str(uriage.loc[uriage['item_name'] == trg]['item_price'].min()))
```

## ノック17:日付の揺れを補正

```python
kokyaku_daicho['登録日'].astype('str')
```

```python
flag_is_serial = kokyaku_daicho['登録日'].astype('str').str.isdigit()
flag_is_serial
```

```python
flag_is_serial.sum()
```

```python
fromSerial = pd.to_timedelta(kokyaku_daicho.loc[flag_is_serial, '登録日'].astype('float'), unit='D') + pd.to_datetime('1900/01/01')
fromSerial
```

```python
fromString = pd.to_datetime(kokyaku_daicho.loc[~flag_is_serial, '登録日'])
fromString
```

```python
kokyaku_daicho['登録日'] = pd.concat([fromSerial, fromString])
kokyaku_daicho.head()
```

```python
kokyaku_daicho['登録年月'] = kokyaku_daicho['登録日'].dt.strftime('%Y%m')
```
